```python
%matplotlib inline
import matplotlib.pyplot as plt

import suncal
```

# Plotting Results

Pre-configured plots may be made using the plot_* methods of the output object.All functions take an optional matplotlib figure argument which can be used to add plot elements after drawing is complete.

Note: The semicolon after the plot function call suppresses the returned figure from being displayed twice in the notebook.

```python
u = suncal.UncertCalc.from_configfile('../ex_magphase.yaml')  # Run an example to plot
u.calculate()
```

## Output Plot Examples

```python
u.out.plot_pdf();
```

```python
u.out.plot_pdf(funcs=[0], intervals=[.68, .90, .95, .99], legend=False, label='desc');
```

```python
u.out.mc.plot_correlation(contour=True, bins=40);
```

## Input Plot Examples

```python
u.out.mc.plot_xhists();
```

```python
u.out.mc.plot_xhists(inpts=[0], bins=400);
```

```python
u.out.mc.plot_xscatter(points=10000);
```

```python
u.out.mc.plot_xscatter(contour=True, bins=40);
```

```python
u.out.mc.plot_converge();
```

## Plot Attributes

Keyword arguments to the plot_XX functions are passed to the underlying matplotlib plot or hist method.

```python
u.out.gum.plot_pdf(ls=':', color='darkgreen')
```

```python
u.out.mc.plot_pdf(bins=200, histtype='step')
```
