### Resistor Combinations

What happens to uncertainty in resistor value by combining resistors in series/parallel combinations?
Keep the total resistance the same, but vary the number of individual components.

Each loop builds not only the nominal values, but also the function used in the calculator.

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

import suncal
```

```python
Rnom = 1000
tol = .05
```

```python
# Parallel combinations
N = 20

p_uncert = np.zeros(N)
nrange = range(1,N+1)
for n in nrange:
    Rn = Rnom*n
    f = '1/(' + '+'.join(['1/R{}'.format(i) for i in range(n)]) + ')'  # Build the function 1/(1/R1 + 1/R2 ... + 1/Rn)
    inputs = [{'name':'R{}'.format(i), 'nom':Rn, 'uncerts':[{'std':Rn*tol}]} for i in range(n)]
    u = suncal.UncertCalc(f, inputs)
    u.calculate(MC=False)
    out = u.out.get_output(method='gum')
    p_uncert[n-1] = out.uncert / out.mean * 100

plt.plot(nrange, p_uncert, marker='.')
plt.xlabel('Parallel #')
plt.ylabel('Uncertainty %')
plt.tight_layout()
```

```python
# Series combinations
M = 20

p_uncert = np.zeros(M)
mrange = range(1,M+1)
for m in mrange:
    Rm = Rnom/m
    f = '+'.join(['R{}'.format(i) for i in range(m)])  # Build the function (R1 + R2 ... + Rm)
    inputs = [{'name':'R{}'.format(i), 'nom':Rm, 'uncerts':[{'std':Rm*tol}]} for i in range(n)]
    u = suncal.UncertCalc(f, inputs)
    u.calculate(MC=False)
    out = u.out.get_output(method='gum')
    p_uncert[m-1] = out.uncert / out.mean * 100

plt.plot(mrange, p_uncert, marker='.')
plt.xlabel('Series #')
plt.ylabel('Uncertainty %')
plt.tight_layout()
```

```python
# Parallel-Series combinations
N = 10  # n parallel resistors
M = 10  # m series groups

p_uncert2 = np.zeros((N,M))
nrange = range(1,N+1)
mrange = range(1,M+1)
for m in mrange:
    for n in nrange:
        Rg = Rnom / m # Each group
        Rm = Rg * n   # Each R in group
        f =  '+'.join([ '1/(' + '+'.join(['1/R{}'.format(i+j*n)  # Build the function (1/(1/R1 + 1/R2 + .. 1/Rn) + 1/(1/R3 + 1/R4 .. + 1/Rn)).. etc
                            for i in range(n)]) +')' for j in range(m) ])
        inputs = [{'name':'R{}'.format(i), 'nom':Rm, 'uncerts':[{'std':Rm*tol}]} for i in range(n*m)]
        u = suncal.UncertCalc(f, inputs)
        u.calculate(MC=False)
        out = u.out.get_output(method='gum')
        p_uncert2[n-1,m-1] = out.uncert / out.mean * 100
```

```python
X, Y = np.meshgrid(nrange,mrange)
plt.contourf(X, Y, p_uncert2, cmap='ocean')
plt.colorbar()
plt.xlabel('Parallel #')
plt.ylabel('Series #')
plt.tight_layout()
```

```python
p = plt.contour(X,Y,p_uncert2, cmap='ocean')
plt.clabel(p, inline=1, fontsize=14)
plt.xlabel('Parallel #')
plt.ylabel('Series #')
plt.tight_layout()
```

```python
plt.pcolor(X, Y, p_uncert2, cmap='ocean_r', edgecolors='black')
plt.colorbar()
plt.xlabel('Parallel #')
plt.ylabel('Series #')
plt.tight_layout()
```
